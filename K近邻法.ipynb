```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsRegressor
```

```python
mcycle = pd.read_csv('C:/Users/Jayson/Desktop/每周代码练习/MLPython_Data/mcycle.csv')
mcycle.shape
```

```python
mcycle.head()
```

```python
sns.scatterplot(x='times', y='accel', data=mcycle)
plt.title('Simulated Motorcycle Accident')
```

```python
X = np.array(mcycle.times).reshape(-1, 1)
```

```python
y = mcycle.accel
```

```python
fig, ax = plt.subplots(2, 2, figsize=(9, 6), sharex=True, sharey=True)
fig.subplots_adjust(hspace=0.5, wspace=0.5)
for i, k in zip([1, 2, 3, 4], [1, 10, 25, 50]):
    model = KNeighborsRegressor(n_neighbors=k)
    model.fit(X, y)
    pred = model.predict(np.arange(60).reshape(-1, 1))
    plt.subplot(2, 2, i)
    sns.scatterplot(x='times',
                    y='accel',
                    s=20,
                    data=mcycle,
                    facecolor='none',
                    edgecolor='k')
    plt.plot(np.arange(60), pred, 'b')
    plt.text(0, 55, f'K = {k}')
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from mlxtend.plotting import plot_decision_regions
```

```python
X, y = load_iris(return_X_y=True)
X2 = X[:, 2:4]
```

```python
fig, ax = plt.subplots(2, 2, figsize=(9,6), sharex=True, sharey=True)
fig.subplots_adjust(hspace=0.1, wspace=0.1)
for i, k in zip([1,2,3,4],[1,10,25,50]):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X2,y)
    plt.subplot(2,2,i)
    plot_decision_regions(X2, y, model)
    plt.xlabel('petal_length')
    plt.ylabel('petal_width')
    plt.text(0.3, 3, f'k = {k}')
plt.tight_layout()
```
